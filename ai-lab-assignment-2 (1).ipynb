```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
# load dataset
digits=load_digits()
#daataset info
print("Dataset Shape:", digits.data.shape)
print("Image  shape: ",digits.images.shape)
print("target label:",digits.target_names)
#
plt.figure(figsize=(4,4))
plt.imshow(digits.images[4],cmap='gray')
plt.title(f"label:{digits.target[0]}")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load dataset
digits = load_digits()
fig, axes = plt.subplots(2, 5, figsize=(7, 5))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Label: {digits.target[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
