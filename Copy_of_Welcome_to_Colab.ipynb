wap in python laxical analyser( scanner ) to identify qr , operator, constant, delimiter, and special symbol


```
keywords = ["int", "float", "char", "if", "else", "return"]
operators = ["+", "-", "*", "/", "="]
delimiters = [";", ",", "(", ")", "{", "}"]
special_symbols = ["#", "<", ">"]

code = input("Enter C Program: ")

for ch in "(){};,+-*/=<>#":
    code = code.replace(ch, " " + ch + " ")

tokens = code.split()

for token in tokens:
    if token in keywords:
        print(token, "-> Keyword")
    elif token in operators:
        print(token, "-> Operator")
    elif token in delimiters:
        print(token, "-> Delimiter")
    elif token in special_symbols:
        print(token, "-> Special Symbol")
    elif token.isdigit():
        print(token, "-> Constant")
    elif token.isalpha():
        print(token, "-> Identifier")
    else:
        print(token, "-> Invalid Token")
```

wap in python that count number of keyword, identifier , operator , delimiter, constant, special symbol in a given c program

```
keywords = ["int", "float", "char", "if", "else", "return"]
operators = ["+", "-", "*", "/", "="]
delimiters = [";", ",", "(", ")", "{", "}"]
special_symbols = ["#", "<", ">"]

code = input("Enter C Program: ")

for ch in "(){};,+-*/=<>#":
    code = code.replace(ch, " " + ch + " ")

tokens = code.split()
keyword_count = 0
operator_count = 0
delimiter_count = 0
special_symbol_count = 0
constant_count = 0
identifier_count = 0
invalid_count = 0
for token in tokens:
    if token in keywords:
        print(token, "-> Keyword")
        keyword_count += 1
    elif token in operators:
        print(token, "-> Operator")
        operator_count += 1
    elif token in delimiters:
        print(token, "-> Delimiter")
        delimiter_count += 1
    elif token in special_symbols:
        print(token, "-> Special Symbol")
        special_symbol_count += 1
    elif token.isdigit():
        print(token, "-> Constant")
        constant_count += 1
    elif token.isalpha():
        print(token, "-> Identifier")
        identifier_count += 1
    else:
        print(token, "-> Invalid Token")
        invalid_count += 1

print("Total Tokens =", len(tokens))
print("Keywords =", keyword_count)
print("Operators =", operator_count)
print("Delimiters =", delimiter_count)
print("Special Symbols =", special_symbol_count)
print("Constants =", constant_count)
print("Identifiers =", identifier_count)
print("Invalid Tokens =", invalid_count)
```

wap in python to removal of comment (single and multiple line comment) from a source program

```


print("Enter C Program (Type END to stop):")

inside_comment = False

while True:
    line = input()

    if line.strip() == "END":
        break

    # Multi-line comment start
    if "/*" in line:
        inside_comment = True
        line = line[:line.index("/*")]

    # Multi-line comment end
    if "*/" in line:
        inside_comment = False
        line = line[line.index("*/") + 2:]

    # Skip lines inside multi-line comments
    if inside_comment:
        continue

    # Remove single-line comment
    if "//" in line:
        line = line[:line.index("//")]

    # Print non-empty line
    if len(line.strip()) > 0:
        print(line)
```
