```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
# load dataset
digits=load_digits()
#daataset info
print("Dataset Shape:", digits.data.shape)
print("Image  shape: ",digits.images.shape)
print("target label:",digits.target_names)
#
plt.figure(figsize=(4,4))
plt.imshow(digits.images[4],cmap='gray')
plt.title(f"label:{digits.target[0]}")
plt.axis('off')
plt.show()
```
